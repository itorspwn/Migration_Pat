#### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## EDA

```python
census_data = pd.read_csv('Resources/census_data.csv')
```

```python
census_data.head()
```

```python
census_data.drop(columns=['Unnamed: 0', 'State'], inplace = True)
```

```python
census_data.head()
```

```python
census_data.info()
```

```python
census_data.describe()
```

```python
ca_data = census_data.loc[census_data['Name'] == 'California']
```

```python
ca_data.head(10)
```

### Employment - CA

```python
employment_data = census_data.filter(['Household Income', 'Median house price', 'Moved to a different state' , 'Poverty Rate', 'Unemployment Rate'], axis=1)
```

```python
employment_data.mean()
```

```python
sns.heatmap(employment_data.corr())
plt.title('Economic Correlation')
plt.savefig("Output/corr_study.png", bbox_inches='tight', dpi = 400)
```

```python
sns.pairplot(employment_data)
```

```python
sns.displot(employment_data['Moved to a different state'])
```

```python
rate_filter_data = ca_data.filter(['Name', 'Year', 'Poverty Rate', 'Unemployment Rate'], axis = 1)
```

```python
rate_data = rate_filter_data.melt(id_vars=['Name', 'Year'], var_name = 'Poverty and Unemployment', value_name = 'Poverty and Unemployment Rates (%)')
```

```python
sns.set_style('darkgrid')

sns.relplot(data=rate_data, x='Year', y='Poverty and Unemployment Rates (%)', hue='Poverty and Unemployment', kind='line')
plt.title('Poverty and Unemployment Rates in California')
plt.savefig("Output/povertyunemployment_rate.png", bbox_inches='tight', dpi = 400)
```

### Gender - CA

```python
gender_filter_data = ca_data.filter(['Name', 'Moved to different state Male', 'Moved to different state Female', 'Year'], axis = 1)
```

```python
gender_data = gender_filter_data.melt(id_vars=['Name', 'Year'], var_name = 'Gender', value_name = 'Number of People')
```

```python
sns.relplot(data=gender_data, x='Year', y='Number of People', hue='Gender', kind='line')
plt.title('Moving Out of California - Male vs. Female')
plt.savefig("Output/male_female.png", bbox_inches='tight', dpi = 400)
```

### Education - CA

```python
ed_filter_data = ca_data.filter(['Name', 'Year', 'Moved to different state Less than high school graduate', 'Moved to different state High school graduate', "Moved to different state Some college or associate's degree", "Moved to different state Bachelor's degree", "Moved to different state Graduate or professional degree"  ], axis = 1)
```

```python
ed_data = ed_filter_data.melt(id_vars=['Name', 'Year'], var_name = 'Level of Education', value_name = 'Number of People')
ed_data.head(10)
```

```python
sns.relplot(data=ed_data, x='Year', y='Number of People', hue='Level of Education', kind='line')
plt.title('Moving Out of California - Education Level')
plt.savefig("Output/education.png", bbox_inches='tight', dpi = 400)
```

### Poverty Level - CA

```python
pl_filter_data = ca_data.filter(['Name', 'Year', 'Moved to different state Below 100 percent of the poverty level', 'Moved to different state 100 to 149 percent of the poverty level', 'Moved to different state above 150 percent of the poverty level'], axis = 1)
```

```python
pl_data = pl_filter_data.melt(id_vars=['Name', 'Year'], var_name = 'Poverty Level', value_name = 'Number of People')
```

```python
sns.relplot(data=pl_data, x='Year', y='Number of People', hue='Poverty Level', kind='line')
plt.title('Moving Out of California - Poverty Level')
plt.savefig("Output/poverty.png", bbox_inches='tight', dpi = 400)
```

### Age Group - CA

```python
age_filter_data = ca_data.filter(['Name', 'Year', 
                                     'Moved to different state 1 to 4 years', 
                                     'Moved to different state 5 to 17 years',
                                     'Moved to different state 18 to 24 years',  
                                     'Moved to different state 25 to 34 years', 
                                     'Moved to different state 35 to 44 years', 
                                     'Moved to different state 45 to 54 years',
                                     'Moved to different state 55 to 64 years',  
                                     'Moved to different state 65 to 74 years',
                                     'Moved to different state 75 years and over' ], axis = 1)
```

```python
age_data = age_filter_data.melt(id_vars=['Name', 'Year'], var_name = 'Age Group', value_name = 'Number of People')
```

```python
sns.relplot(data=age_data, x='Year', y='Number of People', hue='Age Group', kind='line')
plt.title('Moving Out of California - Age Group')
plt.savefig("Output/age.png", bbox_inches='tight', dpi = 400)
```
