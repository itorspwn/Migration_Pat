```python
# dependencies
import pymongo
import pandas as pd
import json
```

```python
# connect to MongoDB using default port
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# declare the database
db = client.migration_db

# declare the collections
census_data = db.census_data
flow_data = db.flow_data
```

```python
# read in our data via csv files
census_df = pd.read_excel("Resources/pooja/census_data.xlsx")
flow_df = pd.read_excel("Resources/pooja/MigrationflowfromCAtootherstates.xls")
```

```python
census_df
```

```python
flow_df
```

```python
# clean up the dataframes for clarity, to insert into database
# remove first index column
census_df = census_df.drop(['Unnamed: 0'], axis = 1)
census_df
```

```python
# rename column names
flow_df.columns = ['current_residence', '2009_estimate', '2009_moe', '2010_estimate', '2010_moe', '2011_estimate', '2011_moe',
                   '2012_estimate', '2012_moe', '2013_estimate', '2013_moe', '2014_estimate', '2014_moe', '2015_estimate',
                   '2015_moe', '2016_estimate', '2016_moe', '2017_estimate', '2017_moe', '2018_estimate', '2018_moe',
                   '2019_estimate', '2019_moe']
```

```python
# delete first row
flow_df = flow_df.drop(flow_df.index[0])
flow_df
```

```python
# convert dataframes into a list of dictionaries
# use .to_dict and 'records' orient to make our desired list
post_census = census_df.to_dict(orient='records')
post_flow = flow_df.to_dict(orient='records')
```

```python
# insert all documents into database collections using .insert_many
census_data.insert_many(post_census)
```

```python
flow_data.insert_many(post_flow)
```

```python
# verify results using .find
```

```python
# query database collections and transform the data back into a pandas dataframe
results1 = census_data.find()
results2 = flow_data.find()

# creates DataFrame.  
df_census = pd.DataFrame(results1)
df_flow = pd.DataFrame(results2)
```

```python
# remove first index column '_id'
df_census = df_census.drop(['_id'], axis = 1)
df_census
```

```python
# remove first index column '_id'
df_flow = df_flow.drop(['_id'], axis = 1)
df_flow
```

```python
# output dataframe from database as a json file to visualize with javascript
df_census.to_json('census_df.json', orient = 'records') 
df_flow.to_json('flow_df.json', orient = 'records') 
```
